```python
import pandas as pd

# read a csv file
data = pd.read_csv("./data/tips.csv")

# print the top 10 rows of data
display(data.head())

# see the rank of the array using `.shape`
data.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# import the data
data = pd.read_csv("./data/tips.csv")

# create a scatter plot
# 1. prep the graph
#      create the type of graph
plt.scatter(data['day'], data['tip'])
    
#      add a title
plt.title("Tip Data")

#      set the labels of the graph
plt.xlabel("Day")
plt.ylabel("Tip ($)")

# 2. show the graph
plt.show()
```

```python
# Add some color to the scatter plot


import pandas as pd
import matplotlib.pyplot as plt

# import the data
data = pd.read_csv("./data/tips.csv")

# create a scatter plot
# 1. prep the graph
#      create the type of graph
#      add information from SIZE of party
#      add information from the amount of
#          the TOTAL BILL
plt.scatter(data['day'], data['tip'], s=data['size'], c=data['total_bill'])
    
#      add a title
plt.title("Tip Data")

#      set the labels of the graph
plt.xlabel("Day")
plt.ylabel("Tip ($)")

#      add a colour gradient to show magnitude of colours
plt.colorbar()

# 2. show the graph
plt.show()
```

```python
# Bar chart

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/tips.csv')

# Create bar chart
plt.bar(data['day'], data['tip'], width=0.4, fill=False)

plt.title("Tips for Each Day")

plt.xlabel("Day")
plt.ylabel("Tip ($)")

plt.show()
```
