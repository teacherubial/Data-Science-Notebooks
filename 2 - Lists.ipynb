## `Set` datatype

A set is a list that contains UNIQUE elements.  
In other words, no repeats.

```python
some_list = [1, 1, 2, 3, 4, 5, 5, 6, 6]

some_set = set(some_list)  # Quick way to remove repeats
some_set
```

```python
# To create a set

some_set = {2, 1, 3, 4, 5}
some_set
```

## Iterating over `lists`
It's really useful to loop through items  
inside lists. You can do that using  
a specific type of `for` loop.

`for` _individual element_ `in` _list_:

```python
list_of_groceries = ["Apples", "Bananas", "Shampoo"]

for item in list_of_groceries:
    print(f"You're going to need {item.lower()}.")
    
set_of_groceries = {
    "apples",
    "chips",
    "junk food"
}

for item in set_of_groceries:
    print(f"You're going to need {item.lower()}.")
```

## List Comprehension

```python
fruits = [
    "apple",
    "banana",
    "cherry",
    "starfruit",
    "mango",
]

# Grab all fruit with letter a in their name
a_fruits = []
for fruit in fruits:
    if "a" in fruit:
        a_fruits.append(fruit)

a_fruits
```

```python
fruits = [
    "apple",
    "banana",
    "cherry",
    "starfruit",
    "mango",
]

# List comprehension
# Grab all the fruits with a in their name
a_fruits = [x for x in fruits if "a" in x]
a_fruits

# Capitalize all the fruits' names
cap_fruits = [x.capitalize() for x in fruits]
cap_fruits
```

## Splitting Strings to Lists

We can use the `.split()` method to split strings to lists.

```python
some_string = "100 100 100 100 100"

split_string = some_string.split(" ")

split_nums = [int(item) for item in split_string]
split_nums
```

## Adding to a `list`
This is the best way to add to a list.  
Use the `.append()` method.

```python
list_of_students = [
    [1, "Mr. Ubial"]
]

print(list_of_students)

list_of_students.append([2, "David M."])
list_of_students += [[3, "John P."]]

print(list_of_students)
```

## Deleting from a `list`
Use the `.pop()` method.

`.pop()` returns the value of the item in the list.

```python
list_of_students = [
    [1, "Mr. Ubial"]
]

print(list_of_students)

list_of_students.append([2, "David M."])

print(list_of_students)

list_of_students.pop(1)   # Return the popped item
print(list_of_students)
```

## Slices of `list`s and `str`s

```python
this_string = "123456789"

# get the right-most character
print(this_string[-1])
print(this_string[-2])

# get the first five characters
print(this_string[:5])
# list[start value:end value:(step)]

# from the 5th character on
print(this_string[5:])

print(this_string[3:6])

# every second character
print(this_string[1::2])
# reverse string
print(this_string[-1::-1])
```
